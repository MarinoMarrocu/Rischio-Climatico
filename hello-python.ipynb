## Un piccolo ripasso

Python è stato chiamato così per la formazione comica britannica Monty Python, quindi faremo del nostro primo programma Python un omaggio al loro sketch sullo Spam.

**Assegnazione variabile**: Qui creiamo una variabile chiamata spam_amount e le assegniamo il valore 0 usando =, che è chiamato operatore di assegnazione.

```python
print(spam_amount)
```

**Chiamata di funzioni**: print è una funzione Python che visualizza il valore passato ad essa sullo schermo. Chiamiamo le funzioni mettendo le parentesi dopo il loro nome, e mettendo gli ingressi (o gli argomenti) alla funzione in quelle parentesi.

```python
# Ordering Spam, egg, Spam, Spam, bacon and Spam (4 more servings of Spam)
spam_amount = spam_amount + 4
```

La prima riga sopra è un commento. In Python, i commenti iniziano con il simbolo #.

Poi vediamo un esempio di riassegnazione. La riassegnazione del valore di una variabile esistente ha lo stesso aspetto della creazione di una variabile - utilizza ancora l'operatore di assegnazione =.

In questo caso, il valore che stiamo assegnando a spam_amount comporta una semplice aritmetica sul suo valore precedente. Quando incontra questa linea, Python valuta l'espressione sul lato destro della = (0 + 4 = 4), e poi assegna quel valore alla variabile sul lato sinistro.

```python
if spam_amount > 0:
    print("But I don't want ANY spam!")

viking_song = "Spam Spam Spam"
print(viking_song)
```

Non parleremo molto di " condizionali " se non più tardi, ma, anche se non avete mai programmato prima d'ora, potete probabilmente indovinare a cosa serve. Python è apprezzato per la sua leggibilità e la semplicità.

Notate come abbiamo indicato quale codice appartiene all'if. "Ma non voglio QUALSIASI spam!" deve essere stampato solo se spam_amount è positivo. Ma il codice successivo (come print(viking_song)) dovrebbe essere eseguito a prescindere. Come facciamo a saperlo (e Python)?

I due punti (:) alla fine della riga if indicano che sta iniziando un nuovo "blocco di codice". Le righe successive che sono rientrate fanno parte di quel blocco di codice. Alcuni altri linguaggi usano {curly braces} per segnare l'inizio e la fine dei blocchi di codice. L'uso di Python di spazi bianchi significativi può essere sorprendente per i programmatori che sono abituati ad altri linguaggi, ma in pratica può portare ad un codice più coerente e leggibile rispetto ai linguaggi che non fanno rispettare l'indentazione dei blocchi di codice.

Le righe successive che trattano di viking_song non sono frastagliate con 4 spazi extra, quindi non fanno parte del blocco di codice dell'if. Vedremo più avanti altri esempi di blocchi di codice indentati quando definiremo le funzioni e useremo i loop.

Questo frammento di codice è anche il nostro primo avvistamento di una stringa in Python:

"Ma io non voglio NESSUNO spam!"
Le stringhe possono essere contrassegnate sia da virgolette doppie che singole. (Ma poiché questa particolare stringa contiene un carattere a virgolette singole, potremmo confondere Python cercando di circondarla con le virgolette singole, a meno che non stiamo attenti).

```python
viking_song = "Spam " * spam_amount
print(viking_song)
```

L'operatore * può essere usato per moltiplicare due numeri (3 * 3 vale 9), ma in modo abbastanza divertente, possiamo anche moltiplicare una stringa per un numero, per ottenere una versione che è stata ripetuta così tante volte. Python offre una serie di trucchetti per risparmiare tempo, come questo, in cui gli operatori come * e + hanno un significato diverso a seconda del tipo di applicazione.

## Quesito 1

Qual'é l'energia cinetica di un gas monoatomico a 21°C?

````python
# K = 3/2 k T
TC = 21 # temperatura in °C
#k = 1.38*10**(-23) # costante di Boltzmann
k = 1.38E-23 # costante di Boltzmann
T = TC + 273.15 # temperatura in Kelvin
K = (3.0/2.0) * k * T
print("Energia cinetica: ",K,"J") # 6.09E-21 ???
````

````python
#N_A = 6.022E23 # Numero di Avogadro
N_A = 6.022*10**23 # Numero di Avogadro

K_mole = N_A * K # Energia di una mole di gas
print("Energia cinetica si una mole: ",K_mole,"J")
````
